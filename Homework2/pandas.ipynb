# Задание 1
Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

ratings = pd.read_csv('ratings.csv')
ratings.head(10)
```

```python
movies = pd.read_csv('movies.csv')
movies.head(10)
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
filtered_joined = joined[(joined ['rating']== 5.0)]
filtered_joined
```

```python
rate = filtered_joined.groupby(['movieId' , 'title'])['movieId'].count()
rate.sort_values(ascending = False).head(1)
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head(10)
```

```python
filtered_countries = data[(data['country']=='Estonia') | (data['country']=='Lithuania') | (data['country']=='Latvia') ]
filtered_countries.head(10)
```

```python
filtered_categories = filtered_countries[(filtered_countries['category']==4) | (filtered_countries['category']==12) | (filtered_countries['category']==21) ]
filtered_categories.head(10)
```

```python
filtered_years = filtered_categories [(filtered_categories['year'] >= 2005)& (filtered_categories['year'] <= 2010)]
filtered_years.head(10)
```

```python
filtered_quantity = filtered_years [(filtered_years['quantity']>0)]
filtered_quantity.head(10)
```

```python
filtered_quantity['quantity'].sum()
```
