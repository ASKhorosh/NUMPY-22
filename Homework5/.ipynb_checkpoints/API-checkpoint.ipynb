## Обязательная часть
Будем парсить страницу со свежеми новостям на habr.com/ru/all/.

Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:

KEYWORDS = ['python', 'парсинг']

Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).

В итоге должен формироваться датафрейм вида: <дата> - <заголовок> - <ссылка>

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
KEYWORDS = ['python', 'парсинг']
```

```python
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
posts = soup.find_all('div', class_='tm-articles-list')
posts
```

```python
list_hubs = pd.DataFrame()
for post in posts:
    hubs = post.find_all('div', class_='tm-article-snippet')
    for hub in hubs:
        hub_lower = hub.text.lower()
        if any([key in hub_lower for key in KEYWORDS]):
            date_element = post.find('span', class_ = 'tm-article-snippet__datetime-published').text
            title_element = post.find('div', class_='tm-article-snippet').text
            link_element = title_element.attrs.get('href').text
            d = {'дата': date_element, 'заголовок': title_element, 'ссылка':link_element}
            list_hubs = pd.concat([list_hubs, pd.DataFrame([d])])
list_hubs
```
